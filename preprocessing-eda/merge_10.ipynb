```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
def total_df(merged, name_str):
  temp = pd.read_csv(f'/content/drive/MyDrive/final_project/{name_str}.csv',index_col=0)
  temp.drop_duplicates(subset='headline',inplace=True)
  return pd.concat([merged,temp])

names = ['samsung','kakao','kb','lgchem','naver','celltrion','hyundai','posco','shinhan','skhynix']
df = pd.DataFrame()
for name in names:
  df = total_df(df, name)

df 
```

```python
df.reset_index(inplace=True)
del df['index']
df
```
